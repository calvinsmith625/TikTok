```
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import FunctionTransformer
```

```
# normally you need to load data into your environment
# usually if it is a csv file you can use pd.read_csv
# pandas is how you convert any data into a workable form
import pandas as pd
# the RandomForestRegressor is what will build our model
# sklearn will do the leg work, as long as you can give it good data
from sklearn.ensemble import RandomForestRegressor
```

```
# these functions are designed to build a pretend dataset

def func_log(x, a, b, c):
    """Return values from a general log function."""
    return a * np.log(b * x) + c

def generate_data(func, *args, jitter=0):
    """Return a tuple of arrays with random data along a general function."""
    xs = np.linspace(1, 50, 500)
    ys = func(xs, *args)
    noise = jitter * np.random.normal(size=len(xs)) + jitter
    xs = xs.reshape(-1, 1)                                 
    ys = (ys + noise).reshape(-1, 1)
    return xs, ys
```

```
# Data
x_data, y_data = generate_data(func_log, 100000, 1, 5000, jitter=30000)            

# Visualization
plt.scatter(x_data, y_data, alpha=.7)            
plt.title("Predict Tiktok Views")
plt.xlabel('Number of Dance Moves')
plt.ylabel('Number of Views')
```

```
model = RandomForestRegressor()
# x_data is the data you're using to predict
# y_data is what you're trying to predict
# in this case the number of views is being predicted by the number of dance moves
model.fit(x_data, y_data)
```

```
model.score(x_data, y_data)
```
